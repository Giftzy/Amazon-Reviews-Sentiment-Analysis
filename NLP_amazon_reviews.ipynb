```python
import pandas as pd
from textblob import TextBlob
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('./excel_work-data/amazon_reviews.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df['reviewText']=df['reviewText'].astype(str).fillna('')
df['reviewText']
```

```python
df['sentiment']=[TextBlob(text).sentiment.polarity for text in df['reviewText']]
df['sentiment']
```

```python
df.tail()
```

```python
categorized_sentiment=[]
for polarity in df['sentiment']:
    if polarity>0:
        categorized_sentiment.append('positive')
    elif polarity<0:
        categorized_sentiment.append('negative')
    else:
        categorized_sentiment.append('neutral')
df['categorized_sentiment']=categorized_sentiment
```

```python
df['categorized_sentiment']
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['categorized_sentiment'],test_size=0.2,random_state=42)
```

```python
predicted_sentiment=[]
for p in X_test:
    polarity=TextBlob(p).sentiment.polarity
    if polarity>0:
        predicted_sentiment.append('positive')
    elif polarity<0:
        predicted_sentiment.append('negative')
    else:
        predicted_sentiment.append('neutral')
```

```python
acc=accuracy_score(y_test,predicted_sentiment)
```
